```python
def mostra_img(imagem):
  fig = plt.gcf()
  fig.set_size_inches(18, 6)
  plt.imshow(imagem, cmap='gray')
  plt.axis('off')
  plt.show()
```

```python
def exibir_limiarizacao(img, limiar = 127):
  _, tresh_binary = cv2.threshold(gray, limiar, 255, cv2.THRESH_BINARY)
  _, tresh_binary_inv = cv2.threshold(gray, limiar, 255, cv2.THRESH_BINARY_INV)
  _, tresh_trunc = cv2.threshold(gray, limiar, 255, cv2.THRESH_TRUNC)
  _, tresh_tozero = cv2.threshold(gray, limiar, 255, cv2.THRESH_TOZERO)
  _, tresh_tozero_inv = cv2.threshold(gray, limiar, 255, cv2.THRESH_TOZERO_INV)

  titulos = ['Imagem Original', 'Binary', 'Binary Inv', 'Trunc', 'To zero', 'To zero inv']
  imagens = [img, tresh_binary, tresh_binary_inv, tresh_trunc, tresh_tozero, tresh_tozero_inv]

  fig = plt.gcf()
  fig.set_size_inches(18,12)

  for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(cv2.cvtColor(imagens[i], cv2.COLOR_BGR2RGB), cmap='gray')
    plt.title(titulos[i])
    plt.xticks([]), plt.yticks([])
  plt.show()
```

```python
exibir_limiarizacao(gray, limiar)
```
